```python
vctk_path = '/content/vctk/'
```

now we've loaded all the spk_embs

```python
import os
import numpy as np
from tqdm import tqdm
```

```python
os.makedirs(vctk_path + 'spk_emb_agg', exist_ok=True)
```

```python
spk_list = os.listdir(vctk_path + 'spk_emb')
```

```python
for spk in tqdm(spk_list):
  spk_path = vctk_path + 'spk_emb/' + spk
  spk_embs = []
  for utt_emb_filename in os.listdir(spk_path):
    spk_embs.append(np.load(spk_path + '/' + utt_emb_filename))
  emb_agg = np.mean(np.stack(spk_embs), axis=0)
  assert emb_agg.shape == (512,)
  np.save(f'{vctk_path}spk_emb_agg/{spk}.npy', emb_agg)
```
